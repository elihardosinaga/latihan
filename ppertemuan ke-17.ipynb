```python
import pandas as pd

data = pd.read_csv("/content/drive/MyDrive/bootcamp data analyst/dataset_final_project.csv")
print(data)
```

```python
data["Branch"] = data["Branch"].replace({"A" : "1", "B" : "2", "C" : "3"})
data["City"] = data["City"].replace({"Yangon" : "1", "Naypyitaw" : "2", "Mandalay" : "3"})
data["Customer type"] = data["Customer type"].replace({"Normal" : "1", "Member" : "2"})
data["Product line"] = data["Product line"].replace({"Electronic accessories" : "1", "Fashion accessories" : "2", "Food and beverages" : "3", "Health and beauty" : "4", "Home and lifestyle" : "5", "Sports and travel" : "6"})
data["Payment"] = data["Payment"].replace({"Cash" : "1", "Credit card" : "2", "Ewallet" : "3"})

print(data)
```

```python
data["Total"] = data["Unit price"] * data["Quantity"]
data["Tax"] = data["Total"] * 0.05
data["Final price"] = data["Total"] - data["Tax"]

print(data)
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
print(data)
```
